```python
!pip install selenium webdriver-manager
!apt-get update
!wget -q -O - https://dl-ssl.google.com/linux/linux_signing_key.pub | apt-key add -
!echo 'deb [arch=amd64] http://dl.google.com/linux/chrome/deb/ stable main' >> /etc/apt/sources.list.d/google-chrome.list
!apt-get update
!apt-get install google-chrome-stable

from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# 1. 드라이버 설정
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument('--disable-gpu')
chrome_options.add_argument('--window-size=1920,1080')
chrome_options.binary_location = '/usr/bin/google-chrome'

driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)

try:
    # 2. 페이지 접속
    url = "https://comic.naver.com/webtoon"
    driver.get(url)
    wait = WebDriverWait(driver, 10)

    # 3. 알려주신 정확한 경로의 'ul' 태그 대기
    ul_selector = "#container > div.ListSpot__spot_wrap--Iko15 > div.content > div > ul"
    target_ul = wait.until(EC.presence_of_element_located((By.CSS_SELECTOR, ul_selector)))

    # 4. 상단 3개 웹툰 요소 가져오기
    items = target_ul.find_elements(By.TAG_NAME, "li")

    print(f"=== 🔍 실시간 신규 웹툰 3개 리포트 ===\n")

# ... (앞부분 드라이버 설정 코드는 동일) ...

    for i, item in enumerate(items[:3], 1):
        raw_text = item.text

        # 1. 줄바꿈으로 쪼개고 깨끗한 줄만 남기기
        lines = [line.strip() for line in raw_text.split('\n') if line.strip()]

        # 2. [수정 포인트] 첫 번째 줄이 '신작'이면 리스트에서 제거하여 밀림 방지
        if lines and lines[0] == "신작":
            lines.pop(0) # '신작' 글자 삭제

        # 3. 이제 다시 순서대로 할당 (제대로 정렬됨)
        if len(lines) >= 3:
            title = lines[0]      # 진짜 제목 (예: 애드립의 신)
            author = lines[1]     # 진짜 작가 (예: GO / 남장현 / 전명)
            description = " ".join(lines[2:]) # 나머지 줄거리 전체
        elif len(lines) == 2:
            title, author = lines[0], lines[1]
            description = "내용 요약이 없습니다."
        else:
            continue

        # 4. 결과 출력
        print(f"[{i}번 웹툰]")
        print(f"📌 제목 : {title}")
        print(f"👤 작가 : {author}")
        print(f"📝 내용 : {description}")
        print("-" * 50)

# ... (뒷부분 마감 코드는 동일) ...

except Exception as e:
    print(f"처리 중 오류가 발생했습니다: {e}")

finally:
    # 5초 뒤 종료
    time.sleep(5)
    driver.quit()
```
