**Why the Memory Addresses are Different:**
- **Temporary Instance (id(Cheff())):**
> Each call to Cheff() creates a new, unique instance of the Cheff class.
When you call id(Cheff()), a new instance is created temporarily for the id function call, and this instance is not stored or referenced after the id call.
Since it is a new instance, it has a unique memory address each time you call Cheff().

- **Persistent Instance (id(user_1)):**
> When you call user_1 = Cheff(), an instance of Cheff is created and stored in the variable user_1.
The id(user_1) call retrieves the memory address of this specific, persistent instance.

```python
class Car:
    # 1. 생성자: 객체가 만들어질 때 호출되는 함수
    def __init__(self, brand, model, color):
        self.brand = brand  # 브랜드
        self.model = model  # 모델명
        self.color = color  # 색상
        self.speed = 0      # 현재 속도 (기본값 0)

    # 2. 메서드: 자동차의 동작 (가속)
    def accelerate(self, amount):
        self.speed += amount
        print(f"{self.model}이(가) {amount}km/h 가속하여 현재 {self.speed}km/h입니다.")

    # 3. 메서드: 자동차의 동작 (정지)
    def stop(self):
        self.speed = 0
        print(f"{self.model}이(가) 멈췄습니다.")
```

```python
# 'my_car'라는 실제 자동차 객체 생성
my_car = Car("Hyundai", "Ionic 6", "White")

# 'your_car'라는 또 다른 자동차 객체 생성
your_car = Car("Tesla", "Model 3", "Black")

# 각 자동차의 정보 출력
print(f"내 차는 {my_car.brand} {my_car.model}입니다.")
# 출력: 내 차는 Hyundai Ionic 6입니다.

# 자동차 동작시키기
my_car.accelerate(50)  # Ionic 6이 50km/h로 달립니다.
your_car.accelerate(80) # Model 3는 80km/h로 달립니다.
```

## **과제 1. faker를 사용하여 1000명의 데이터를 class를 이용하여 만들어라**

- [click faker](https://faker.readthedocs.io/en/master/)

```python
from faker import Faker
import pandas as pd

fake = Faker('ko_KR') # Use Korean locale for more realistic data if preferred, or remove for default English

class User:
    def __init__(self, name, age, address, phone_number, job):
        self.name = name
        self.age = age
        self.address = address
        self.phone_number = phone_number
        self.job = job

    def __repr__(self):
        return f"User(Name: {self.name}, Age: {self.age}, Job: {self.job})"

# Generate 1000 fake users
users_data = []
for _ in range(1000):
    name = fake.name()
    age = fake.random_int(min=18, max=90)
    address = fake.address()
    phone_number = fake.phone_number()
    job = fake.job()
    users_data.append(User(name, age, address, phone_number, job))

print(f"Generated {len(users_data)} user instances.")

# Convert to pandas DataFrame
df_users = pd.DataFrame([
    {
        'Name': user.name,
        'Age': user.age,
        'Address': user.address,
        'Phone Number': user.phone_number,
        'Job': user.job
    } for user in users_data
])

print("\nFirst 5 rows of the generated DataFrame:")
display(df_users.head())
```

## **과제 2. Tom이 Jerry를 쫒아가는 행위를 method(함수)로 만들어 보자(10번 following)**

### **class User:**
> **def __init __ (self):**
>   - ##### Initialize attributes
>   - **HAS (Attiributes)**

> **def following_mode():**
>   - ##### change attributes (following, follower) for chasing
>   - **Does (Method, Change the attributes)**

<img src= 'https://pbs.twimg.com/profile_images/810136393/tom-jerry-pic_400x400.jpg' width=300>

```python
class Character:
    def __init__(self, name):
        self.name = name
        self.following = [] # To keep track of who this character is following

    def follow(self, target_character):
        if target_character not in self.following:
            self.following.append(target_character)
        print(f"{self.name} is chasing {target_character.name}!")

# Create Tom and Jerry instances
tom = Character("Tom")
jerry = Character("Jerry")

# Simulate Tom following Jerry 10 times
print("--- Tom chasing Jerry (10 times) ---")
for i in range(1, 11):
    tom.follow(jerry)
    # Optionally, you could add a delay here to make it more like a chase
    # import time
    # time.sleep(0.5)

print(f"\nTom is now following: {[c.name for c in tom.following]}")
```
